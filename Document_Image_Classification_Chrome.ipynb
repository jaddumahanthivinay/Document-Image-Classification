```python
import tensorflow as tf
import os
import numpy as np
import pandas as pd

df = pd.read_csv('labels_final.csv',dtype=str)
```

```python
df.head()
```

```python
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Dropout, Flatten, Dense
from tensorflow.keras import applications
from keras_preprocessing.image import ImageDataGenerator
from keras.layers import Input
from keras.layers import Dense,Reshape
from keras.layers import Flatten
from keras.layers import Embedding
from keras.models import Model
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.layers.merge import concatenate
from keras.models import Model
from keras.layers import Dropout
from keras.models import Sequential
```

```python
image_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    width_shift_range=0.15,
    height_shift_range=0.15,
    shear_range=0.1,
    zoom_range=0.1,
    validation_split=0.20
)
```

```python
train_generator = image_gen.flow_from_dataframe(
    dataframe=df,
    directory="./data_final/",
    x_col='path',
    y_col='label',
    target_size=(224, 224),
    class_mode='categorical',
    batch_size=32,
    shuffle=True,
    seed=1,
    subset='training'
)
valid_generator = image_gen.flow_from_dataframe(
    dataframe=df,
    directory="./data_final/",
    x_col='path',
    y_col='label',
    target_size=(224, 224),
    class_mode='categorical',
    batch_size=32,
    shuffle=True,
    seed=1,
    subset='validation'
)
```

```python
input_tensor = Input(shape=(224, 224, 3))
model = applications.ResNet50V2(include_top=False,weights='imagenet',input_tensor=input_tensor)
layer_dict = dict([(layer.name, layer) for layer in model.layers])
x = layer_dict['post_relu'].output
conv_layer_1 = Conv2D(filters=64, kernel_size=(3,3), padding='same',activation='relu')(x)
pool_layer_1 = MaxPooling2D(pool_size=(2,2),padding='same')(conv_layer_1)
flatten = Flatten()(pool_layer_1)
FC1 = Dense(128, activation='relu')(flatten)
FC2 = Dense(64, activation='relu')(FC1)
output = Dense(16, activation='softmax')(FC2)
model_final = Model(input_tensor,output)
for layer in model_final.layers[:190]:
    layer.trainable = False
model_final.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model_final.summary()
```

```python
STEP_SIZE_TRAIN=train_generator.n//train_generator.batch_size
STEP_SIZE_VALID=valid_generator.n//valid_generator.batch_size
model_final.fit_generator(generator=train_generator,
                    steps_per_epoch=STEP_SIZE_TRAIN,
                    validation_data=valid_generator,
                    validation_steps=STEP_SIZE_VALID,
                    epochs=10
)
```

```python
tf.keras.utils.plot_model(model_final,to_file='first_model.png')
```

```python
input_tensor = Input(shape=(224, 224, 3))
model= applications.ResNet50V2(include_top=False, weights='imagenet',input_tensor=input_tensor)
conv_layer_1 = Conv2D(filters=128, kernel_size=(7,7), padding='valid',activation='relu')(model.output)
conv_layer_2 = Conv2D(filters=64, kernel_size=(1,1), padding='same',activation='relu')(conv_layer_1)
# flatten=Flatten()(conv2)
res = Reshape((-1,))(conv_layer_2)
output_layer = Dense(16,activation='softmax')(res)
model_final = Model(inputs=[input_tensor], outputs=output_layer)
for layer in model_final.layers[:190]:
    layer.trainable = False
model_final.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model_final.summary()
```

```python
STEP_SIZE_TRAIN=train_generator.n//train_generator.batch_size
STEP_SIZE_VALID=valid_generator.n//valid_generator.batch_size
model_final.fit_generator(generator=train_generator,
                    steps_per_epoch=STEP_SIZE_TRAIN,
                    validation_data=valid_generator,
                    validation_steps=STEP_SIZE_VALID,
                    epochs=10
)
```
